```python
import ast

def parse_specifications(spec_string):
    # Step 1: Replace the "=>" with ":"
    spec_string = spec_string.replace("=>", ":")
    
    # Step 2: Convert the string to a dictionary using ast.literal_eval
    try:
        spec_dict = ast.literal_eval(spec_string)
        spec_list = spec_dict.get('product_specification', [])
        
        # Step 3: Extract key-value pairs and create a string
        spec_str = " ".join([f"{item['key']}: {item['value']}," for item in spec_list if 'key' in item and 'value' in item])
    except (ValueError, SyntaxError):
        spec_str = ""
    
    return spec_str

# Apply the parsing function to the dataset
df2['parsed_specifications'] = df2['product_specifications'].apply(parse_specifications)

# Display the parsed specifications for verification
parsed_specs_sample = df2['parsed_specifications'].head(1)
parsed_specs_sample.tolist()
```

```python
def clean_category_tree(category_tree):
    # Remove brackets and split the tree into categories
    categories = category_tree.strip("[]").split(" >> ")
    return " ".join(categories)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load mBERT model and tokenizer
model_name = "bert-base-multilingual-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

def get_embedding(text, tokenizer, model):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
    return embeddings.squeeze().numpy()

# Generate embeddings for the combined text
df2['embedding'] = df2['combined_text'].apply(lambda x: get_embedding(x, tokenizer, model))
```

```python
df2['embedding']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search_products(query, df, tokenizer, model, top_k=5):
    # Generate embedding for the search query
    query_embedding = get_embedding(query, tokenizer, model)

    #print(query_embedding)
    # Compute cosine similarity between the query and product embeddings
    similarities = []
    for idx, product_embedding in enumerate(df['embedding']):
        similarity = cosine_similarity([query_embedding], [product_embedding]).item()
        similarities.append((df.iloc[idx], similarity))
    
    # Sort products by similarity score
    sorted_products = sorted(similarities, key=lambda x: x[1], reverse=True)
    
    # Return the top-k products
    top_products = sorted_products[:top_k]
    return top_products

# Example search query
query = "baby"
top_products = search_products(query, df2, tokenizer, model)

# Display the results
for product, score in top_products:
    print(f"Product: {product[['product_name','discounted_price','description']]}, Similarity Score: {score:.4f}")
```
